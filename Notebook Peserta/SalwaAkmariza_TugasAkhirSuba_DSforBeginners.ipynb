```python
# Import library
import pandas as pd
import numpy as np

# Eksekusi syntax untuk mengcustom tampilan
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', '{:2f}'.format)
```

```python
# Ekstrasi data
df = '/content/StudentMentalHealth_dirty.csv'
data_SMH = pd.read_csv(df)

# Menampilkan hasil
display(data_SMH.head())
```

```python
# Informasi umum pada data
data_SMH.info()
```

```python
# Melihat deskripsi data
data_SMH.describe()
```

```python
# Ubah kolom age dan screen time menjadi angka
data_SMH['Age'] = pd.to_numeric(data_SMH['Age'], errors='coerce')
data_SMH['Screen Time (hrs/day)'] = pd.to_numeric(data_SMH['Screen Time (hrs/day)'], errors='coerce')
```

```python
# Memperbaiki data pada kolom Age, ubah outlier menjadi NaN
data_SMH.loc[data_SMH['Age'] > 100, 'Age'] = np.nan

# Memperbaiki data pada kolom Screen Time (hrs/day)
data_SMH.loc[data_SMH['Screen Time (hrs/day)'] < 0, 'Screen Time (hrs/day)'] = np.nan
data_SMH.loc[data_SMH['Screen Time (hrs/day)'] > 15, 'Screen Time (hrs/day)'] = np.nan

# Memperbaiki data pada kolom Sleep Duration
data_SMH.loc[data_SMH['Sleep Duration (hrs)'] < 0, 'Sleep Duration (hrs)'] = np.nan
data_SMH.loc[data_SMH['Sleep Duration (hrs)'] > 24, 'Sleep Duration (hrs)'] = np.nan

# Memperbaiki data pada kolom Physical Activity (hrs/week), misal seminggu max 21 jam
data_SMH.loc[data_SMH['Physical Activity (hrs/week)'] < 0, 'Physical Activity (hrs/week)'] = np.nan
data_SMH.loc[data_SMH['Physical Activity (hrs/week)'] > 21, 'Physical Activity (hrs/week)'] = np.nan
```

```python
# Mengganti semua data yang NaN menjadi median data
data_ganti = ['Age', 'Screen Time (hrs/day)', 'Sleep Duration (hrs)', 'Physical Activity (hrs/week)']
data_SMH[data_ganti] = data_SMH[data_ganti].fillna(data_SMH[data_ganti].median())

# Melihat deskripsi
data_SMH.describe()
```

```python
# Cek sisa data kosong
data_SMH.isna().sum()
```

```python
# Mengisi data kosong pada kolom Nama
data_SMH['Name'].fillna('Unknown', inplace=True)

# Menggabungkan data yang diisi modus
data_modus = data_SMH[['Gender', 'Education Level', 'Stress Level', 'Anxious Before Exams', 'Academic Performance Change']]

# Mengisi data_modus pakai modus
for col in data_modus:
  modus = data_modus[col].mode()[0]
  data_SMH[col].fillna(modus, inplace=True)

# Mengecek data bersih
data_SMH.isna().sum()
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

# Melihat sebaran tingkat stress
sn.countplot(x='Stress Level', data=data_SMH)
plt.show()
```

```python
# Melihat hubungan Education Level dengan Sleep Duration
sn.boxplot(x='Education Level', y='Sleep Duration (hrs)', data=data_SMH)
```

```python
plt.figure(figsize=(12, 8))
# Mencari hubungan Screen Time dengan Sleep Duration beserta garis regresi
sn.regplot(x='Screen Time (hrs/day)', y='Sleep Duration (hrs)', data=data_SMH)
```

```python
# Melihat hubungan Sleep Duration (hrs) dengan Level Stress
sn.boxplot(x='Sleep Duration (hrs)', y='Stress Level', data=data_SMH)
```

```python
# Mencari hubungan Sleep Duration dengan Anxious Before Exams
sn.boxplot(x='Sleep Duration (hrs)', y='Anxious Before Exams', data=data_SMH)
plt.title('Hubungan Sleep Duration dengan Anxious Before Exams pada Pelajar')
```

```python
# Hubungan Sleep Duration dan Anxious Before Exams dengan Stress Level
sn.boxplot(x='Sleep Duration (hrs)', y='Anxious Before Exams', hue='Stress Level', data=data_SMH)
plt.show()
```
