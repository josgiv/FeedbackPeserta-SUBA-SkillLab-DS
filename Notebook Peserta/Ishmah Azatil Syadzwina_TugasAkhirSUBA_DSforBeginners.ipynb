### **Gunakan Teknik Winsorizing untuk mengatasi Outlier**
Teknik Winsorizing digunakan untuk mengatasi outlier dengan membatasi nilai ekstrem pada batas persentil tertentu. Pendekatan ini bertujuan untuk mengurangi pengaruh nilai ekstrem terhadap hasil analisis tanpa menghilangkan data, sehingga distribusi data menjadi lebih stabil dan representatif.

```python
# Fungsi untuk teknik winsorizing
def teknik_winsorizing(series):
    # Hitung Q1, Q3, dan IQR
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    # Hitung lower dan upper bound
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Jika lower bound negatif maka ubah menjadi 0
    if(lower_bound < 0):
      lower_bound = 0

    # Winsorizing: clip nilai ke batas bawah & atas
    series = series.astype(pd.Float64Dtype())
    winsorized_series = series.clip(lower=lower_bound, upper=upper_bound)

    return (winsorized_series)
```

```python
for col in kolom_numerik:
    student_mental_health[col] = teknik_winsorizing(student_mental_health[col])
    box_plot(student_mental_health[col], col, '#B07AA1')
```
