# A.1.1 Data Dictionary
| Kolom                          | Deskripsi                                                                    |
|:-------------------------------|:-----------------------------------------------------------------------------|
| **Name**                       | Berisi nama-nama responden (student) yang mengisi survei.      |
| **Gender**                     | Berisi jenis kelamin responden yakni 'Male' atau 'Female'. |
| **Age**                        | Berisi usia responden dalam format angka.                  |
| **Education Level**            | Menjelaskan tentang jenjang pendidikan responden, misalnya 'Class 8', 'MSc', 'BTech', 'BA', atau 'Class 11'. |
| **Screen Time (hrs/day)**      | Waktu layar (dalam jam per hari) responden dalam menghabiskan waktu di depan layar Handphone atau Elektronik lainnya. |
| **Sleep Duration (hrs)**       | Berisi durasi tidur (dalam jam) responden. |
| **Physical Activity (hrs/week)** | Aktivitas fisik (dalam jam per minggu) yang dilakukan oleh responden. |
| **Stress Level**               | Menjelaskan tentang tingkat stres dalam bentuk level sesuai dengan yang dialami oleh responden. |
| **Anxious Before Exams**       | Tentang kecemasan responden sebelum melakukan ujian. |
| **Academic Performance Change**| Tentang bagaimana perubahan kinerja akademik dari responden. |
