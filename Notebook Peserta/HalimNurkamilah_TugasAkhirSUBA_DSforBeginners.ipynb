### Visualizing Correlation Matrix with a Heatmap

To better understand the relationships between numerical variables, we'll create a heatmap of the correlation matrix. A heatmap uses color intensity to represent the correlation coefficient between each pair of variables, making it easy to spot strong positive (closer to 1), strong negative (closer to -1), and weak (closer to 0) correlations.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(data_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Group by Age and calculate the mean of Screen Time and Sleep Duration
age_trends = mental_health.groupby('Age').agg(
    avg_screen_time=('Screen Time (hrs/day)', 'mean'),
    avg_sleep_duration=('Sleep Duration (hrs)', 'mean')
).reset_index()

# Create a line plot
fig = go.Figure()

fig.add_trace(go.Scatter(x=age_trends['Age'], y=age_trends['avg_screen_time'],
                         mode='lines+markers',
                         name='Average Screen Time (hrs/day)',
                         line=dict(color='blue')))

fig.add_trace(go.Scatter(x=age_trends['Age'], y=age_trends['avg_sleep_duration'],
                         mode='lines+markers',
                         name='Average Sleep Duration (hrs)',
                         line=dict(color='green')))

fig.update_layout(title='Average Screen Time and Sleep Duration by Age',
                  xaxis_title='Age',
                  yaxis_title='Hours',
                  hovermode='x unified')

fig.show()
```

### Interpretasi Diagram Garis: Rata-rata Waktu Layar dan Durasi Tidur berdasarkan Usia

Diagram garis di atas membandingkan bagaimana rata-rata waktu yang dihabiskan di depan layar (Screen Time) dan rata-rata durasi tidur (Sleep Duration) berubah seiring dengan bertambahnya usia peserta survei.

*   **Rata-rata Waktu Layar (Garis Biru):** Kita dapat melihat fluktuasi rata-rata waktu layar di berbagai kelompok usia. Ada beberapa puncak dan lembah, yang menunjukkan bahwa waktu layar tidak selalu meningkat atau menurun secara linier dengan usia.
*   **Rata-rata Durasi Tidur (Garis Hijau):** Serupa dengan waktu layar, durasi tidur juga menunjukkan pola yang berfluktuasi. Ini menunjukkan bahwa durasi tidur rata-rata juga bervariasi di berbagai kelompok usia.

**Perbandingan dan Implikasi:**

Secara umum, kedua garis cenderung bergerak dalam rentang nilai yang relatif stabil, meskipun ada beberapa anomali atau fluktuasi. Tidak ada tren naik atau turun yang sangat kuat dan konsisten untuk kedua variabel seiring bertambahnya usia. Ini mendukung temuan dari analisis korelasi sebelumnya, yang menunjukkan bahwa hubungan linear antara variabel-variabel numerik ini sangat lemah atau tidak ada.

Namun, diagram ini memungkinkan kita untuk mengidentifikasi kelompok usia tertentu yang mungkin memiliki rata-rata waktu layar yang lebih tinggi atau durasi tidur yang lebih rendah, yang bisa menjadi area untuk penyelidikan lebih lanjut terkait dampaknya terhadap kesehatan mental atau prestasi akademik.

```python
# Convert 'Stress Level' to numerical values for aggregation
stress_level_mapping = {'Low': 0, 'Med': 1, 'Medium': 1, 'High': 2}
mental_health['Stress Level_Numeric'] = mental_health['Stress Level'].map(stress_level_mapping)

age_trends = mental_health.groupby('Age').agg(
    avg_screen_time=('Screen Time (hrs/day)', 'mean'),
    avg_stress_level=('Stress Level_Numeric', 'mean') # Use the new numeric column
).reset_index()

# Create a line plot
fig = go.Figure()

fig.add_trace(go.Scatter(x=age_trends['Age'], y=age_trends['avg_screen_time'],
                         mode='lines+markers',
                         name='Average Screen Time (hrs/day)',
                         line=dict(color='blue')))

fig.add_trace(go.Scatter(x=age_trends['Age'], y=age_trends['avg_stress_level'],
                         mode='lines+markers',
                         name='Average Stress Level (numeric)', # Update name for clarity
                         line=dict(color='red')))

fig.update_layout(title='Average Screen Time and Stress Level by Age',
                  xaxis_title='Age',
                  yaxis_title='Hours / Stress Level (numeric)', # Update y-axis label
                  hovermode='x unified')

fig.show()
```

### Interpretasi Diagram Garis: Rata-rata Waktu Layar dan Tingkat Stres berdasarkan Usia

Diagram garis ini menampilkan bagaimana rata-rata waktu yang dihabiskan di depan layar (Screen Time) dan rata-rata tingkat stres (Stress Level, dalam skala numerik) berubah seiring bertambahnya usia peserta survei.

*   **Rata-rata Waktu Layar (Garis Biru):** Garis biru menunjukkan fluktuasi rata-rata waktu layar di berbagai kelompok usia. Kita bisa melihat ada beberapa puncak dan lembah, yang menandakan bahwa waktu layar tidak secara konsisten meningkat atau menurun seiring usia. Sebagai contoh, ada beberapa usia di mana waktu layar rata-ratanya lebih tinggi atau lebih rendah dibandingkan usia sekitarnya.

*   **Rata-rata Tingkat Stres (Garis Merah):** Garis merah menggambarkan rata-rata tingkat stres yang juga berfluktuasi antar kelompok usia. Seperti waktu layar, tingkat stres tidak menunjukkan tren linear yang jelas (selalu naik atau selalu turun) seiring bertambahnya usia.

**Perbandingan dan Implikasi:**

Secara umum, kedua garis (waktu layar dan tingkat stres) cenderung bergerak dalam rentang nilai yang relatif stabil, meskipun ada variasi di setiap usia.
```python
# import library imblearn
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import OrdinalEncoder

# Impute NaN values in 'Age' column for both training and testing sets
# Calculate median from the training set to prevent data leakage
median_age_train = x_train['Age'].median()
x_train['Age'] = x_train['Age'].fillna(median_age_train)
x_test['Age'] = x_test['Age'].fillna(median_age_train)

# Drop 'Name' column from x_train and x_test as it's an identifier and not numerical
if 'Name' in x_train.columns:
    x_train = x_train.drop(columns=['Name'])
if 'Name' in x_test.columns:
    x_test = x_test.drop(columns=['Name'])

# Identify remaining categorical columns in x_train
categorical_cols_for_encoding = x_train.select_dtypes(include='object').columns

# Apply OrdinalEncoder to the categorical columns in x_train and x_test
ord_enc = OrdinalEncoder()
x_train[categorical_cols_for_encoding] = ord_enc.fit_transform(x_train[categorical_cols_for_encoding])
x_test[categorical_cols_for_encoding] = ord_enc.transform(x_test[categorical_cols_for_encoding]) # Use transform for test set

# Explicitly cast all numerical columns to float64 to
# prevent casting issues with imblearn
# Select all numeric columns for casting (now including the newly encoded ones)
numeric_cols_to_cast = x_train.select_dtypes(include=['number']).columns
x_train[numeric_cols_to_cast] = x_train[numeric_cols_to_cast].astype('float64')
x_test[numeric_cols_to_cast] = x_test[numeric_cols_to_cast].astype('float64')

# buat object SMOTE
smote = SMOTE()

# terapkan pada X_train dan x_test dan buat variabel baru hasil oversampling
x_train_resample, y_train_resample = smote.fit_resample(x_train, y_train)
```
